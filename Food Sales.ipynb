```python
# Option 1: Fill missing ratings with the mean
order_details['Rating'] = order_details['Rating'].fillna(order_details['Rating'].mean())
```

```python
user_details.drop_duplicates(inplace=True)
cooking_sessions.drop_duplicates(inplace=True)
order_details.drop_duplicates(inplace=True)
```

```python
print(user_details.isnull().sum())
print(cooking_sessions.isnull().sum())
print(order_details.isnull().sum())
```

```python
# Merge user_details with cooking_sessions
merged_data = pd.merge(cooking_sessions, user_details, on='User ID', how='inner')

# Merge the above result with order_details
final_data = pd.merge(merged_data, order_details, on=['User ID', 'Session ID'], how='inner')

# Display the merged data
final_data.head()
```

```python
final_data.columns
```

```python
# Drop duplicate columns (e.g., '_y' version of the duplicate columns)
merged_df = final_data.drop(columns=['Dish Name_y', 'Meal Type_y'])

# Optionally, rename '_x' columns to remove suffix
final_dataset = merged_df.rename(columns={
    'Dish Name_x': 'Dish Name',
    'Meal Type_x': 'Meal Type'
})

```

## User Behavior and Demographics

### Location-based Analysis

```python
# Plot orders by location
location_counts = final_dataset['Location'].value_counts()
location_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Orders by Location')
plt.xlabel('Location')
plt.ylabel('Number of Orders')
plt.show()
```

## Cooking Session Analysis

### Most Popular Meal Types

```python
# Plot meal type popularity
meal_counts = final_dataset['Meal Type'].value_counts()
meal_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Most Popular Meal Types')
plt.xlabel('Meal Type')
plt.ylabel('Number of Sessions')
plt.show()
```

### Session Duration

```python
# Box plot for session durations
sns.boxplot(x=final_dataset['Duration (mins)'])
plt.title('Session Duration Distribution')
plt.xlabel('Session Duration (mins)')
plt.show()
```

### Session Ratings

```python
# Average rating by meal type
ratings_by_meal = final_dataset.groupby('Meal Type')['Session Rating'].mean()
ratings_by_meal.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Average Session Ratings by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Average Rating')
plt.show()
```

### Time of Day for Cooking Sessions

```python
# Heatmap for time of day and day of the week
heatmap_data = final_dataset.pivot_table(index=final_dataset['Session Start'].dt.day_name(),
                                columns=final_dataset['Time of Day'],
                                values='Session ID', aggfunc='count')
sns.heatmap(heatmap_data, cmap='Blues', annot=True)
plt.title('Cooking Sessions by Time of Day and Day')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
plt.show()
```

### Order Trends Analysis

```python
# Top dishes ordered
dish_counts = final_dataset['Dish Name'].value_counts().head(10)
dish_counts.plot(kind='bar', color='red', edgecolor='black')
plt.title('Top Dishes Ordered')
plt.xlabel('Dish Name')
plt.ylabel('Number of Orders')
plt.show()
```

### Order Status Distribution

```python
# Order status distribution
status_counts = final_dataset['Order Status'].value_counts()
status_counts.plot(kind='pie', autopct='%1.1f%%', colors=['green', 'red', 'yellow'])
plt.title('Order Status Distribution')
plt.show()
```

### Amount Spent Per User

```python
# Box plot for amount spent
sns.boxplot(x=final_dataset['Amount (USD)'])
plt.title('Amount Spent per User')
plt.xlabel('Amount (USD)')
plt.show()
```

### Correlation Insights

### Cooking Session Rating vs Order Rating

```python
#Scatter plot for session rating vs order rating
sns.scatterplot(x=final_dataset['Session Rating'], y=final_dataset['Rating'])
plt.title('Session Rating vs Order Rating')
plt.xlabel('Session Rating')
plt.ylabel('Order Rating')
plt.show()
```

### Session Duration vs Order Amount

```python
# Scatter plot for session duration vs order amount
sns.scatterplot(x=final_dataset['Duration (mins)'], y=final_dataset['Amount (USD)'])
plt.title('Session Duration vs Order Amount')
plt.xlabel('Session Duration (mins)')
plt.ylabel('Order Amount (USD)')
plt.show()
```

### Seasonal Trends

#### Seasonal Meal Preferences

```python
# Line chart for meal preferences by month
final_dataset['Month'] = final_dataset['Order Date'].dt.month
seasonal_pref = final_dataset.groupby(['Month', 'Meal Type'])['Order ID'].count().unstack()
seasonal_pref.plot(kind='line', marker='o', colormap='viridis')
plt.title('Seasonal Meal Preferences')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()
```

### Order Trends by Date

```python
# Line chart for order trends by date
daily_orders = final_dataset.groupby('Order Date')['Order ID'].count()
daily_orders.plot(kind='line', color='blue', marker='o')
plt.title('Order Trends Over Time')
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.show()
```

### Session rating Vs Order amount

```python
# Relationship between session rating and order amount
sns.scatterplot(data= final_dataset, x="Session Rating", y="Amount (USD)", hue="Meal Type")
plt.title("Session Rating vs Order Amount")
plt.savefig("rating_vs_amount.png")
plt.show()
```

### Popular dishes

```python
# Popular dishes
popular_dishes = final_dataset["Dish Name"].value_counts().head(10)
print(popular_dishes)
popular_dishes.plot(kind="bar", color="teal", title="Top 10 Popular Dishes")
plt.xlabel("Dish Name")
plt.ylabel("Number of Orders")
plt.savefig("popular_dishes.png")
plt.show()
```
